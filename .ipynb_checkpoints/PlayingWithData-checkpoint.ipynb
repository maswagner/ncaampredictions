```python
curdata = pd.read_csv('TeamPoints.csv')
```

```python
plt.scatter(curdata['PPG'], curdata['OPP PPG'])
plt.xlabel('PPG')
plt.ylabel('OPP PPG')
```

```python
curdata['PTDif'] = curdata['PPG'] - curdata['OPP PPG']
```

```python
curdata['Team'] = curdata['Team'].str.replace(' ', '')
curdata['Team'] = curdata['Team'].str.lower()
curdata
```

```python
curdata[curdata['Team'] == 'arizona']
```

```python
team1 = input('Select a team: ')
team1 = team1.lower()

def findteam(team):
    plt.figure(figsize=(12,10))

    for i in range(1,10):
        curdata['RandomScore'] = curdata['PPG']+random.randint(-10,10)
        curdata['OPPRandomScore'] = curdata['OPP PPG']+random.randint(-10,10)
        plt.scatter(i, (curdata[curdata['Team'] == team]['RandomScore']), c='darkblue')
        plt.scatter(i, (curdata[curdata['Team'] == team]['OPPRandomScore']), c='darkred')
    plt.axhline(y = float(curdata[curdata['Team'] == team]['PPG']), label='Average PPG')
    plt.axhline(y = float(curdata[curdata['Team'] == team]['OPP PPG']), color='red', label='Average OPP PPG')
    plt.title(team.title() + ' Random Points over 10 games')
    plt.legend()
    plt.xlabel('Number of games played')
    plt.ylabel('Total points scored by '+ team.title())

findteam(team1)
```
